```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.array([5,6,7,8,9,10])
y=np.array([18,20,22,25,28,32])
plt.scatter(x,y,40,color='r',label="Scatter_data",marker='x')
plt.xlabel('age')
plt.ylabel('weight')
plt.title("age_weight_prediction")
plt.grid(True ,color='k')
plt.show()
```

```python
# mean values
mean_x=np.mean(x)
mean_y=np.mean(y)
n=len(x)
```

```python
# to get the values for m and c
num=0 
den=0
for i in range (n):
    num=num+((x[i]-mean_x)*(y[i]-mean_y))
    den=den+((x[i]-mean_x)**2)
m=num/den
#y=mx+c,c=y-mx
c=mean_y-(m*mean_x)
print(m,c)
```

```python
#calculate y prediction
y_pred=np.array([])     #to create empty array

for i in range(n):
    p=(m*x[i])+c #y=mx+c y=p
    y_pred=np.append(y_pred,np.array([p]),axis=0)  #to add data into empty array using numpy
    
print(y_pred)    
```

```python
#plot the regression line -predicted data
plt.plot(x,y_pred,color='m',label='regression_line')
#plot the actual data
plt.scatter(x,y,40,color='g',label="predicted data",marker='x') #predicted data
plt.xlabel("age")
plt.ylabel("weight")
plt.legend()
plt.show()
```

```python
# how good our fit line orregression line by using r square method
nu=0
de=0
for i in range(n):
    nu=nu+((y_pred[i]-mean_y)**2)    #predicted value of y
    de=de+((y[i]-mean_y)**2)         #actual values of y
r2=nu/de
r2

#r2 ranges from 0 to 1 if r2>0.7 then its good fit line
```
